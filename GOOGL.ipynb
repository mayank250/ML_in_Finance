```python
import pandas as pd
from pandas import DatetimeIndex
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid', context='talk', palette='Dark2')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import math
import seaborn as sns
sns.set(style='darkgrid', context='talk', palette='Dark2')
my_year_month_fmt = mdates.DateFormatter('%m/%y')
plt.rcParams['figure.figsize'] = (15, 9)
%matplotlib inline
import pandas_datareader as web
import datetime
```

```python
start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2019, 12, 31)
googl_df = web.DataReader('GOOGL', 'yahoo', start, end)
googl = googl_df.copy()
```

# 20, 100 Days SMA

Here I took 20 days for the short term investors and 100 days longer term investors

```python
googl_df["20d_sma"]=np.round(googl_df["Close"].rolling(window=20,center=False).mean(),5)
googl_df["100d_sma"]=np.round(googl_df["Close"].rolling(window=100,center=False).mean(),5)
googl_df.fillna(0,inplace=True)
googl_df.tail(20)
```

```python
start_date = '2019-01-01'
end_date = '2020-01-01'

fig, ax = plt.subplots(figsize=(15,10))

ax.plot(googl_df.loc[start_date:end_date, :].index, googl_df.loc[start_date:end_date, 'Close'], label='Actual Price', color = 'red')
ax.plot(googl_df.loc[start_date:end_date, :].index, googl_df.loc[start_date:end_date, '20d_sma'], label = '20-days SMA Span', color = 'green')
ax.plot(googl_df.loc[start_date:end_date, :].index, googl_df.loc[start_date:end_date, '100d_sma'], label = '100-days SMA Span', color = 'orange')

ax.legend(loc='best')
ax.set_title('SMA-20,100 days')
ax.set_xlabel('Obsevation Date')
ax.set_ylabel('Price in $')
```

# 20, 100 Days EMA

```python
ema_20= googl.ewm(span=20, adjust=False).mean()
ema_100 = googl.ewm(span=100, adjust=False).mean()

fig, ax = plt.subplots(figsize=(15,10))

ax.plot(googl.loc[start_date:end_date, :].index, googl.loc[start_date:end_date, 'Close'], label='Actual Price', color = 'red')
ax.plot(ema_20.loc[start_date:end_date, :].index, ema_20.loc[start_date:end_date, 'Close'], label = '20-days EMA Span', color = 'green')
ax.plot(ema_100.loc[start_date:end_date, :].index, ema_100.loc[start_date:end_date, 'Close'], label = '100-days EMA Span', color = 'orange')

ax.legend(loc='best')
ax.set_title('EMA-20,100 days')
ax.set_xlabel('Obsevation Date')
ax.set_ylabel('Price in $')
```

```python
ema_10 = googl.ewm(span=10, adjust=False).mean()
ema_30 = googl.ewm(span=30, adjust=False).mean()

fig, ax = plt.subplots(figsize=(17,10))

ax.plot(googl.loc[start_date:end_date, :].index, googl.loc[start_date:end_date, 'Close'], label='Actual Price', color = 'red')
ax.plot(ema_10.loc[start_date:end_date, :].index, ema_10.loc[start_date:end_date, 'Close'], label = '10-days EMA Span', color = 'green')
ax.plot(ema_20.loc[start_date:end_date, :].index, ema_20.loc[start_date:end_date, 'Close'], label = '20-days EMA Span', color = 'orange')
ax.plot(ema_30.loc[start_date:end_date, :].index, ema_30.loc[start_date:end_date, 'Close'], label = '30-days EMA Span', color = 'blue')

ax.legend(loc='best')
ax.set_title('EWMA - 10, 20, 30 Days Span')
ax.set_xlabel('Obsevation Date')
ax.set_ylabel('Price in $')
ax.xaxis.set_major_formatter(my_year_month_fmt)
```

# 20 Days EMA vs 20 Days SMA

```python
fig, ax = plt.subplots(figsize=(15,10))

ax.plot(googl.loc[start_date:end_date, :].index, googl.loc[start_date:end_date, 'Close'], label='Actual Price', color = 'red')
ax.plot(ema_20.loc[start_date:end_date, :].index, ema_20.loc[start_date:end_date, 'Close'], label = '20-days EMA Span', color = 'green')
ax.plot(googl_df.loc[start_date:end_date, :].index, googl_df.loc[start_date:end_date, '20d_sma'], label = '20-days SMA Span', color = 'orange')

ax.legend(loc='best')
ax.set_title('20days SMA v.s. 20days EMA')
ax.set_xlabel('Obsevation Date')
ax.set_ylabel('Price in $')
```
